```python
df = df.drop("spotify_url",axis=1)
df = df.drop("album_id",axis=1)
display(df)
```

Colunas:

Clean_name: nome do artista

Album: título do álbum

Rank_2003: posição que o álbum ocupava no ranking de álbuns em 2003

Rank_2012: posição que o álbum ocupava no ranking de álbuns em 2012

Rank_2020: posição que o álbum ocupava no ranking de álbuns em 2020

Differential: a diferença entre o rank de 2003 e 2020, que é negativa se o álbum subiu de posição ao longo dos anos e positiva se desceu

Release_year: ano em que o álbum foi lançado

Genre: gênero musical do álbum 

Type: tipo de álbum, como um álbum de estúdio, uma compilação, uma trilha sonora ou um álbum ao vivo

Weeks_on_billboard: número de semanas que o álbum passou na Billboard 200, que mede o desempenho dos álbuns em termos de vendas e popularidade

Peak_billboard_position: posição mais alta que o álbum alcançou na Billboard 200, que indica o pico de popularidade do álbum

Spotify_popularity: medida de popularidade do álbum no Spotify

Artist_member_count: número de integrantes do grupo musical (para artistas solo, esse valor é 1)

Artist_gender: gênero dos membros do grupo ou do artista solo

Artist_birth_year_sum: soma dos anos de nascimento de todos os membros do grupo ou, no caso de um artista solo, o ano de nascimento desse artista. Esse valor pode ser usado para calcular a idade média dos membros.

Debut_album_release_year: ano em que o artista ou banda lançou seu álbum de estreia

Ave_age_at_top_500: idade média dos membros do grupo (ou do artista solo) no momento em que o álbum foi incluído em uma lista de "Top 500" álbuns

Years_between: diferença entre o ano de lançamento do álbum de estreia do artista e o e o ano em que o álbum foi listado no "Top 500"
