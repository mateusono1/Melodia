```python
df = df.drop("spotify_url",axis=1)
df = df.drop("album_id",axis=1)
display(df)
```

Colunas:

Clean_name: nome do artista

Album: ttulo do lbum

Rank_2003*: posio que o lbum ocupava no ranking de lbuns em 2003

Rank_2012*: posio que o lbum ocupava no ranking de lbuns em 2012

Rank_2020*: posio que o lbum ocupava no ranking de lbuns em 2020

Differential: a diferena entre o rank de 2003 e 2020, que  negativa se o lbum desceu de posio ao longo dos anos e positiva se subiu

Release_year: ano em que o lbum foi lanado

Genre: gnero musical do lbum 

Type: tipo de lbum, como um lbum de estdio, uma compilao, uma trilha sonora ou um lbum ao vivo

Weeks_on_billboard: nmero de semanas que o lbum passou na Billboard 200, que mede o desempenho dos lbuns em termos de vendas e popularidade

Peak_billboard_position: posio mais alta que o lbum alcanou na Billboard 200, que indica o pico de popularidade do lbum

Spotify_popularity: medida de popularidade do lbum no Spotify

Artist_member_count: nmero de integrantes do grupo musical (para artistas solo, esse valor  1)

Artist_gender: gnero dos membros do grupo ou do artista solo

Artist_birth_year_sum: soma dos anos de nascimento de todos os membros do grupo ou, no caso de um artista solo, o ano de nascimento desse artista. Esse valor pode ser usado para calcular a idade mdia dos membros.

Debut_album_release_year: ano em que o artista ou banda lanou seu lbum de estreia

Ave_age_at_top_500*: idade mdia dos membros do grupo (ou do artista solo) no momento em que o lbum foi includo em uma lista de "Top 500" lbuns

Years_between*: diferena entre o ano de lanamento do lbum de estreia do artista e o e o ano em que o lbum foi listado no "Top 500"
