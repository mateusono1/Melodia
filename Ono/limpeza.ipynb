# Exemplo de plotagem de gráficos

```python
a=df.groupby(by="genre", as_index=False)["spotify_popularity"].mean()
a = a.sort_values(by="spotify_popularity", ascending=False)
```

```python
sns.barplot(x="genre", y="spotify_popularity", data=a)



plt.title("Popularidade média por gênero")
plt.ylabel("Popularidade média")
plt.xlabel('Gênero')
plt.xticks(rotation=90)
plt.show()
```

<p> voces estao vendo o que eu estou vendo? exitem duas colunas diferentes para 'blues/blues rock"! Isso provavelmente aconteceu devido a um erro de digitacao e vamos consertar isso: </p>

```python
for col in df.columns:
  print(col.upper())
  print(df[col].unique(), df[col].dtype, "\n")
```

<p> Vou criar tambem um grafico mostrando a quantidade de um certo genero nos rankings de cada ano para melhor vizualizacao </p>

```python
df_melted = pd.melt(df, id_vars=['genre'], 
                    value_vars=['rank_2003', 'rank_2012', 'rank_2020'], 
                    var_name='year', value_name='rank')
df_melted = df_melted.dropna(subset=['rank'])
plt.figure(figsize=(12, 6))
sns.countplot(data=df_melted, x='genre', hue='year')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Genre')
plt.ylabel('Number of Albums in Rank')
plt.title('Number of Albums per Genre in The Rankings of Each Year')
plt.show()
#obg chat gpt :)
```
